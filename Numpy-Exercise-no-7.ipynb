# Complex Numbers

### Q1. Return the angle of a in radian.

```python
a = 1 + 1j
output = np.angle(a,deg = False)
print(output)
```

### Q2. Return the real part and imaginary part of a.

```python
a = np.array([[1+2j],[3+4j],[5+6j]])
real = a.real
imag = a.imag
print("real:",real,"\n")
print("imag:",imag)
```

```python
a = np.array([1+2j,3+4j,5+6j])
a.real = 9
a.imag = [5,7,9]
print(a)
```

# Discrete Fourier Transform

### Q5. Compuete the one-dimensional DFT of a.

```python
a = np.exp(2j * np.pi * np.arange(8))

b = np.fft.fft(a)
print(b)
```

### Q6. Compute the one-dimensional inverse DFT of the output in the above question.

```python
print("a :-\n",a,"\n")
inversed = np.fft.ifft(b)
print("inversed:-",a)
```

```python
a = [0,1,0,0]
b = np.fft.rfft(a)
print(b)

assert b.size == len(a) // 2+1 if len(a)%2 == 0 else len(a)+1//2

b1 = np.fft.fft(a)
print(b1)
```

# Window Functions

```python
fig = plt.figure(figsize = (15,12))

window = np.hamming(51)
plt.plot(np.bartlett(51),label = "Barlett window")
plt.plot(np.blackman(51),label = "Blackman window")
plt.plot(np.hamming(51),label = "Hamming window")
plt.plot(np.hanning(51),label = "Hanning window")
plt.plot(np.kaiser(51,14),label = "Kaiser window")
plt.xlabel('sample')
plt.ylabel('amplitude')
plt.legend()
plt.grid()
plt.show()
```
